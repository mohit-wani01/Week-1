```python
# Scatter plot: Nitrogen (N) vs Rainfall
plt.figure(figsize=(8, 6))
plt.scatter(crop1['N'], crop1['rainfall'], alpha=0.6, color='blue')
plt.title('Nitrogen Content vs Rainfall')
plt.xlabel('Nitrogen Content (N)')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()

```

```python
# Histogram: Humidity distribution
plt.figure(figsize=(8, 6))
plt.hist(crop1['humidity'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Humidity Distribution')
plt.xlabel('Humidity (%)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
